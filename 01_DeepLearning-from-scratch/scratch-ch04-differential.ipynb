## 수치미분을 구해보자 

```python
def numerial_diff(f, x):
    h = 1e-4 # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)
```

* 이처럼 아주 작은 차분으로 미분하는 것을 **수치 미분**라고 한다.
* 한편 수식을 전개해 미분하는 것은 **해석적(Analytic)**라는 말을 이용하여 '해석적 해' 혹은 '해석적으로 미분하다'등으로 표현한다.

* (옮긴이의 설명) 간단히 말해 '해석적 미분'은 우리가 수학시간에 배운 바로 그 미분이고, '수치 미분'은 이를 '근사치'로 계산하는 방법이다. 
* e.g 수치해석학은 "해석학 문제에서 수치적인 근삿값을 구하는 알고리즘을 연구하는 학문이다"

### 4.3.2 수치 미분의 예
* 앞 절의 수치 미분을 사용하여 간단한 함수를 미분해보자

```python
def function_1(x):
    return 0.01*x**2 + 0.1*x
```

```python
# 이 함수를 한 번 그려보자

import numpy as np
import matplotlib.pylab as plt

x = np.arange(0.0, 20.0, 0.1)  # 0에서 20까지 0.1간격의 배열 x를 만든다
y = function_1(x)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y)
plt.show()
```

```python
numerial_diff(function_1, 5)
```

```python
numerial_diff(function_1, 10)
```

```python
numerial_diff(function_1, 2)
```
