## CUDA Tensors

<code>.to</code> ë©”ì†Œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ Tensorë¥¼ ì–´ë– í•œ ì¥ì¹˜ë¡œë„ ì˜®ê¸¸ ìˆ˜ ìˆë‹¤.

ì†ê¸€ì”¨ ë°ì´í„° CUDAë¡œ í•™ìŠµì‹œì¼œë³¼ê¹Œ?<br/>
CPUë¡œ 2ë ˆì´ì–´ ConVëŒë¦¬ëŠ”ë°ë„ ì§„ì§œ í˜ë“¤ì–´í•´ì„œ ì•ˆì“°ëŸ¬ì› ì–´...

```python
# ì´ ì½”ë“œëŠ” CUDAê°€ ì‚¬ìš© ê°€ëŠ¥í•œ í™©ê²½ì—ì„œë§Œ ì‹¤í–‰ëœë‹¤.
# 'torch.device'ë¥¼ ì‚¬ìš©í•˜ì—¬ tensorë¥¼ GPUì•ˆíŒìœ¼ë¡œ ì´ë™í•œë‹¤.

import torch
torch.cuda.is_available()
```

'True'ë¡œ CUDA ì‚¬ìš©ì´ ê°€ëŠ¥í•˜ê¸°ì— CUDAë¡œ Tensorë¥¼ í•™ìŠµì‹œì¼œë³´ì.<br>

cuda deviceë¥¼ ì„ ì–¸í•œë‹¤. => device = <code>torch.device("cuda")</code>
1. <code>to(device)</code>
2. <code>torch.**(, device = device)</code>

```python
if torch.cuda.is_available():
    device = torch.device("cuda")          
    y = torch.ones_like(x, device=device)  # 1. CUDA ì¥ì¹˜ ê°ì²´(device object)ë¡œ GPUìƒì— ì§ì ‘ì ìœ¼ë¡œ tensorë¥¼ ìƒì„±í•˜ëŠ” ë°©ë²•
    x = x.to(device)                       # 2. .to("cuda")ë¥¼ ì‚¬ìš©í•˜ë©´ ëœë‹¤.
    z = x + y
    print(z)
    print(z.to("cpu", torch.double))       # .toëŠ” dtypeë„ í•¨ê»˜ ë³€ê²½í•´ì¤€ë‹¤.
    
else:
    print("OOOOOOOOOOOOOOOOOps!!")
```

ì–´ì œ ì‚¬ìš©í•œ ë˜ ë‹¤ë¥¸ ë°©ë²•<br>
<code>.cuda()</code>

```python
if torch.cuda.is_available():
    device = torch.device("cuda")
    
    y = torch.ones_like(x).cuda()
    x = x.to(device)
    z = y.add_(x)
    print(z)
    print(z.to("cpu"), torch.double)
else:
    print("OOOpps!")
```
