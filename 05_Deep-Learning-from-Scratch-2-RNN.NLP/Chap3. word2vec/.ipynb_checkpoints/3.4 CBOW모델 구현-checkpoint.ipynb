```python
import sys
sys.path.append('..')
import numpy as np
from common.layers import MatMul, SoftmaxWithLoss

class SimpleCBOW:
    def __init__(self, vocab_size, hidden_size):  # ì´ˆê¸°í™” ë©”ì„œë“œì˜ ì¸ìˆ˜ë¡œ vocab_size(ì–´íœ˜ ìˆ˜)ì™€ ì€ë‹‰ì¸µì˜ ë‰´ëŸ° ìˆ˜(hidden_size)ë¥¼ ë°›ëŠ”ë‹¤.
        V, H = vocab_size, hidden_size
        
        # ê°€ì¤‘ì¹˜ ì´ˆê¸°í™”
        W_in = 0.01 * np.random.randn(V, H).astype('f')  # ë¬´ì‘ìœ„ Weight ì´ˆê¸°í™”, íƒ€ì…ì€ float 32ë¹„íŠ¸ë¡œ ë§Œë“¤ì–´ì¤€ë‹¤.
        W_out = 0.01 * np.random.randn(H, V).astype('f')
        
        # ê³„ì¸µ ìƒì„±
        self.in_layer0 = MatMul(W_in)
        self.in_layer1 = MatMul(W_in)  # ì…ë ¥ì—ì„œ ì‚¬ìš©í•˜ëŠ” MatMul ê³„ì¸µì˜ ìˆ˜ëŠ” ì…ë ¥ì— ì‚¬ìš©ë˜ëŠ” ë‹¨ì–´ì˜ ìˆ˜ë§Œí¼ ë§Œë“¤ì–´ì•¼ í•œë‹¤.
                                       # ìœˆë„ìš°ì˜ í¬ê¸°ê°€ 2ì´ë¯€ë¡œ 2ê°œì˜ MatMulê³„ì¸µì„ ë§Œë“¤ì–´ì¤€ë‹¤.
                                       # ìœˆë„ìš°ì˜ í¬ê¸°ê°€ nì´ë¼ë©´ nê°œì˜ MatMulê³„ì¸µì„ ë§Œë“¤ì–´ì£¼ëŠ” ê²ƒì´ë‹¤.
        self.out_layer = MatMul(W_out)
        self.loss_layer = SoftmaxWithLoss()
        
        # ëª¨ë“  ê°€ì¤‘ì¹˜ì™€ ê¸°ìš¸ê¸°ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ëª¨ì•„ì¤€ë‹¤.
        layers = [self.in_layer0, self.in_layer1, self.out_layer]
        self.params, self.grads = [], []  # ì‹ ê²½ë§ì—ì„œ ì‚¬ìš©ë˜ëŠ” ë§¤ê°œë³€ìˆ˜ì™€ ê¸°ìš¸ê¸°ë¥¼ ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ì¸ paramsì™€ grads ë¦¬ìŠ¤íŠ¸ì— ê°ê° ëª¨ì•„ë‘”ë‹¤.
        
        for layer in layers:  
            self.params += layer.params
            self.grads += layer.grads        
        
        # ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ì— ë‹¨ì–´ì˜ ë¶„ì‚° í‘œí˜„ì„ ì €ì¥
        self.word_vecs = W_in
        
    # ìˆœì „íŒŒ forward ë©”ì„œë“œ êµ¬í˜„ 
    def forward(self, contexts, target):
        h0 = self.in_layer0.forward(contexts[:, 0])  # contextsëŠ” 3ì°¨ì› ë°°ì—´ì¸ë°...
        h1 = self.in_layer1.forward(contexts[:, 1])
        h = (h0 + h1) * 0.5
        score = self.out_layer.forward(h)
        loss = self.loss_layer.forward(score, target)
        return loss
    
    # ì—­ì „íŒŒ
    def backward(self, dout=1):
        ds = self.loss_layer.backward(dout)
        da = self.out_layer.backward(ds)
        da *= 0.5
        self.in_layer0.backward(da)
        self.in_layer1.backward(da)
        return None
```
