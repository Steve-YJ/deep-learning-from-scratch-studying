ë‹¤ìŒìœ¼ë¡œ corpusë¡œ ë¶€í„° ë§¥ë½ê³¼ íƒ€ê²Ÿì„ ë§Œë“¤ì–´ë‚¸ë‹¤

* copursë¡œë¶€í„° ë§Œë“  ë§¥ë½ê³¼ íƒ€ê²Ÿ
    * ë§¥ë½ì€ 2ì°¨ì› ë°°ì—´ì´ë‹¤.
    * ë§¥ë½ì˜ 0ë²ˆì§¸ ì°¨ì›ì—ëŠ” ê° ë§¥ë½ ë°ì´í„°ê°€ ì €ì¥ëœë‹¤.
    * ì •í™•í•˜ê²Œ ë§í•˜ë©´, context[0]ì—ëŠ” 0ë²ˆì§¸ ë§¥ë½ì´ ì €ì¥ë˜ê³ , context[1]ì—ëŠ” 1ë²ˆì§¸ ë§¥ë½ì´ ì €ì¥ëœë‹¤.

**ë§¥ë½ê³¼ íƒ€ê²Ÿì„ êµ¬í•˜ëŠ” í•¨ìˆ˜ êµ¬í˜„**

```python
import numpy as np

def create_contexts_target(corpus, window_size=1):
    target = corpus[window_size:-window_size]  # ì¬ë¯¸ìˆëŠ” ì¸ë±ì‹± í‘œí˜„ì´ë‹¤.
    contexts = []
    
    for idx in range(window_size, len(corpus)-window_size):
        cs = [] # cs?
        for t in range(-window_size, window_size + 1):
            if t == 0:
                continue
            cs.append(corpus[idx + t])
        contexts.append(cs)
        
    return np.array(contexts), np.array(target)
        
```

```python
contexts, target = create_contexts_target(corpus, window_size=1)

print('context: ', contexts)
print('target: ', target)
```
